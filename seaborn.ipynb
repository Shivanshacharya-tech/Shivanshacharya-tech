```python
import seaborn as sns
import matplotlib.pyplot as plt
data= sns.load_dataset("fmri")
sns.lineplot(data=data, x="timepoint", y="signal", hue="event", style="region")
plt.title("fmri")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.get_dataset_names()
sns.load_dataset("diamonds")
data=sns.load_dataset("diamonds")
sns.scatterplot(data=data, x="carat", y="price", hue="color")
plt.title("Diamonds: Carat vs Price")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.get_dataset_names()
sns.load_dataset("penguins")
data=sns.load_dataset("penguins")
sns.scatterplot(data=data, x="flipper_length_mm", y="body_mass_g", hue="species")
plt.title("Penguins: Flipper Length vs Body Mass")
plt.show()
```
